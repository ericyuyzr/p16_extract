```python
import re
import numpy as np
import pandas as pd
import spacy
import scispacy
```

```python
raw_data = pd.read_csv('test sample - 20260211.csv')
toy_200 = raw_data.loc[:199, ["msg_id", "anonymized_message"]]
```

```python
def extract_p16_sentence(text):
    # \s{2,} captures spaces, tabs, and newlines (2 or more)
    # The flags make it case-insensitive and allow it to cross line breaks
    pattern = r"\s{2,}(?:(?!\s{2,}).)*p16(?:(?!\s{2,}).)*\s{2,}"
    
    match = re.search(pattern, text, flags=re.IGNORECASE | re.DOTALL)
    
    if match:
        return match.group(0)
    return "No match found"
```

```python
raw_data["p16_sentence"] = raw_data["anonymized_message"].apply(extract_p16_sentence)
```

```python
def clean_clinical_text(text):
    # 1. Replace multiple horizontal spaces/tabs with a single space
    # [ \t]+ matches one or more spaces or tabs but ignores newlines
    step1 = re.sub(r'[ \t]+', ' ', text)
    
    # 2. Clean up "Artifact Clutter" (like those HL7 delimiters ~\&)
    step2 = re.sub(r'[~\\&]', '', step1)
    
    # 3. Remove leading/trailing spaces from each individual line
    cleaned_lines = [line.strip() for line in step2.splitlines()]
    
    # 4. Join back together, removing completely empty lines if desired
    return "\n".join([l for l in cleaned_lines if l])
```

```python
toy_200["p16_clean"] = toy_200["anonymized_message"].apply(clean_clinical_text)
```

```python
def extract_p16_info(text):
    text = text.lower()
    
    # Match any sentence containing "p16" (until the next period)
    pattern = r"p16\s+(?:immunohistochemistry\s+is\s+)?(\w+)"
    
    match = re.search(pattern, text)
    if match:
        return match.group(1)  # entire sentence
    return None

toy_200["p16_info"] = toy_200["p16_clean"].apply(extract_p16_info)
```

```python
nlp = spacy.load("en_core_sci_md")

def extract_medical_entities(text):
    if not isinstance(text, str) or text.strip() == "":
        return []
    
    doc = nlp(text)
    # Extract text and the internal label
    return [(ent.text, ent.label_) for ent in doc.ents]


raw_data['entities'] = raw_data['p16_sentence'].apply(extract_medical_entities)

#toy_200['entity_list'] = toy_200['entities'].apply(lambda x: [item[0] for item in x])

#print(toy_200[['p16_sentence', 'entity_list']].head())
```

```python
def check_p16(entities):
    pattern = re.compile(r'\b(p16|e6h4)\b', re.IGNORECASE)

    for e in entities:
        if isinstance(e, (list, tuple)) and len(e) > 0:
            if pattern.search(str(e[0])):
                return "Found"
    return "Not Found"

raw_data['p16_presence'] = raw_data['entities'].apply(check_p16)
```

```python

def classify_p16(text):
    text_lower = text.lower()

    positive_pattern = r'\bp16\b.{0,30}\b(positive|positivity)\b|\b(positive|positivity)\b.{0,30}\bp16\b'
    negative_pattern = r'\bp16\b.{0,30}\b(negative|negativity|no expression|loss of expression|lack of staining)\b|\b(negative|negativity|no expression|loss of expression|lack of staining)\b.{0,30}\bp16\b'

    # Check negative first
    neg_match = re.search(negative_pattern, text_lower)
    if neg_match:
        return "Negative", text[neg_match.start():neg_match.end()]

    pos_match = re.search(positive_pattern, text_lower)
    if pos_match:
        return "Positive", text[pos_match.start():pos_match.end()]

    return "Neither", None


raw_data[['classification', 'p16_context']] = raw_data['anonymized_message'].apply(
    lambda x: pd.Series(classify_p16(x))
)
```

```python

def classify_p16_with_window(text, window=30):
    text_lower = text.lower()
    
    # Basic patterns
    p16_pattern = re.compile(r'\bp16\b')
    positive_pattern = re.compile(r'\b(positive|positivity)\b')
    
    # Expanded negative pattern: explicit negatives OR negated positivity
    # Includes phrases like "no evidence of p16 positivity"
    neg_p16_pattern = re.compile(
        r'\b('
        r'negative|negativity|no expression|loss of expression|lack of staining|'
        r'no positive|not positive|no positivity|not positivity|'
        r'absence of p16 positivity|lack of p16 positivity|no evidence of p16 positivity'
        r')\b',
        re.IGNORECASE
    )
    
    classification = "Neither"
    context = None
    
    # Check for matches
    p16_matches = list(p16_pattern.finditer(text_lower))
    pos_matches = list(positive_pattern.finditer(text_lower))
    neg_matches = list(neg_p16_pattern.finditer(text_lower))
    
    if not p16_matches:
        return classification, context
    
    closest_dist = float('inf')
    closest_type = None
    closest_span = None
    
    # Loop through all p16 matches
    for p_match in p16_matches:
        p_start, p_end = p_match.span()
        
        # Check positive matches
        for m in pos_matches:
            m_start, m_end = m.span()
            dist = min(abs(m_start - p_end), abs(p_start - m_end))
            
            # Only consider positive if no negation in the ±window context
            snippet_window = text_lower[max(p_start - window, 0):min(m_end + window, len(text_lower))]
            if re.search(r'\b(no|not|lack|absence)\b', snippet_window):
                continue
            
            if dist <= window and dist < closest_dist:
                closest_dist = dist
                closest_type = "Positive"
                closest_span = (min(p_start, m_start), max(p_end, m_end))
        
        # Check negative matches (including negated positives)
        for m in neg_matches:
            m_start, m_end = m.span()
            dist = min(abs(m_start - p_end), abs(p_start - m_end))
            
            if dist <= window and dist < closest_dist:
                closest_dist = dist
                closest_type = "Negative"
                closest_span = (min(p_start, m_start), max(p_end, m_end))
    
    # Return final classification and context snippet
    if closest_type:
        classification = closest_type
        start = max(closest_span[0] - window, 0)
        end = min(closest_span[1] + window, len(text))
        context = text[start:end]
    
    return classification, context

# Apply to your DataFrame
raw_data[['classification', 'p16_context']] = raw_data['anonymized_message'].apply(
    lambda x: pd.Series(classify_p16_with_window(x))
)
```

```python
def find_p16_with_no(text, window=30):
    """
    Returns list of context snippets where 'p16' has the word 'no'
    within +/- window characters.
    """
    text_lower = text.lower()
    
    p16_pattern = re.compile(r'\bp16\b')
    no_pattern = re.compile(r'\bno\b')
    
    results = []
    
    # Find all match spans
    p16_matches = list(p16_pattern.finditer(text_lower))
    no_matches = list(no_pattern.finditer(text_lower))
    
    for p_match in p16_matches:
        p_start, p_end = p_match.span()
        
        for n_match in no_matches:
            n_start, n_end = n_match.span()
            
            # Check if "no" is within +/- window of p16
            if abs(n_start - p_start) <= window:
                
                # Expand snippet for review
                start = max(min(p_start, n_start) - window, 0)
                end = min(max(p_end, n_end) + window, len(text))
                
                snippet = text[start:end]
                results.append(snippet)
                
                break  # Avoid duplicate snippets per p16
    
    return results

raw_data['no_info'] = raw_data['anonymized_message'].apply(find_p16_with_no)
```

```python

def get_balanced_sample(df, label_col='classification', total_n=1000):
    groups = df[label_col].unique()
    n_groups = len(groups)
    
    # Calculate base size per group
    base_n = total_n // n_groups
    remainder = total_n % n_groups
    
    samples = []
    for i, group in enumerate(groups):
        # Add 1 to the sample size for the first few groups to handle remainders
        current_n = base_n + (1 if i < remainder else 0)
        
        # Get all rows for this group
        group_df = df[df[label_col] == group]
        
        # Sample from this group (use replace=True if a group has fewer than 333 rows)
        group_sample = group_df.sample(n=min(len(group_df), current_n), random_state=42)
        samples.append(group_sample)
    
    # Combine and shuffle the final list
    subtable = pd.concat(samples).sample(frac=1, random_state=42).reset_index(drop=True)
    return subtable

# 2. Execute
subtable = get_balanced_sample(raw_data)

# 3. Verify the counts
print(subtable['label'].value_counts())

# 4. Save to CSV
subtable.to_csv('balanced_p16_subtable.csv', index=False)
```
